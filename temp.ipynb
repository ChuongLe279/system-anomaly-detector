```python
import psutil
import time
```

```python
N = 1000
for i in range(N):
    metrics = {
        "timestamp": time.time(),
        "cpu_percent": psutil.cpu_percent(),
        "mem_percent": psutil.virtual_memory().percent,
        "bytes_sent": psutil.net_io_counters().bytes_sent,
        "bytes_recv": psutil.net_io_counters().bytes_recv,
        "disk_read": psutil.disk_io_counters().read_bytes, # type: ignore
        "disk_write": psutil.disk_io_counters().write_bytes, # type: ignore
        "process_count": len(psutil.pids()),
    }
    print(metrics)
    time.sleep(3)
```
